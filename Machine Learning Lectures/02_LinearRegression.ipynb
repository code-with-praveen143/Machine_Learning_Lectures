```python
#Importing the dependencies
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Loading the dataset into the pandas dataframe

df=pd.read_csv("E:\\Datasets\\salary_data.csv")
```

```python
df.head()
df.isnull().sum()
```

```python
#Splitting the data into features and target variables

X=df.iloc[:,:-1].values
Y=df.iloc[:,1].values
```

```python
#print("feature variables are :",X)
#print("target variables are :",Y)
```

```python
#Splitting the dataset into the training data and testing data

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

```python
#Fitting Simple Linear Regression model

from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
model=LinearRegression()
model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
print("My Y prediction is :",y_pred)
 
```

```python
#Visualizing the Trianing results

plt.scatter(X_train,Y_train,color="red")
plt.plot(X_train,model.predict(X_train),linewidth=0.5,color='blue')
plt.xlabel("Years Of Experience")
plt.ylabel("Salary")
plt.title("Salary Vs Experience",color="red")
plt.show()
```

```python
#Visualizing the Testing results

plt.scatter(X_test,Y_test,color="red")
plt.plot(X_test,model.predict(X_test),linewidth=0.5,color='blue')
plt.xlabel("Years Of Experience")
plt.ylabel("Salary")
plt.title("Salary Vs Experience",color="red")
plt.show()
```
